```python
df['booking_date'] = pd.to_datetime(df['booking_date'])
df['month_year'] = df['booking_date'].dt.to_period('M')
monthly_pax_sum = df.groupby('month_year')['pax'].sum().reset_index()
print(monthly_pax_sum)
```

```python
plt.figure(figsize=(10,6))
plt.plot(monthly_pax_sum['month_year'].astype(str), monthly_pax_sum['pax'], marker='o', color='b', linestyle='-')
plt.title('Monthly Pax Count Sum', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Pax Count', fontsize=12)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
total_rows = len(df)
print(f"Total row count: {total_rows}")
count_yes = df[df['refund_status'] == 'Yes'].shape[0]
print(f"Total count with refund_status = 'Yes': {count_yes}")
count_no = df[df['refund_status'] == 'No'].shape[0]
print(f"Total count with refund_status = 'No': {count_no}")
counts = df['refund_status'].value_counts()

# Create bar graph
plt.figure(figsize=(8, 5))
plt.bar(counts.index, counts.values, color=['skyblue', 'salmon'])
plt.xlabel('Refund Status')
plt.ylabel('Count')
plt.title('Count of Refund Status')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
total_rows = len(df)
print(f"Total row count: {total_rows}")
count_yes = df[df['journey_type'] == 'round trip'].shape[0]
print(f"Total count with journey_type = 'round trip': {count_yes}")
count_no = df[df['journey_type'] == 'one way'].shape[0]
print(f"Total count with journey_type = 'one way': {count_no}")
counts = df['journey_type'].value_counts()

# Create bar graph
plt.figure(figsize=(8, 5))
plt.bar(counts.index, counts.values, color=['skyblue', 'salmon'])
plt.xlabel('Journey Types')
plt.ylabel('Count')
plt.title('Count of Journey Types')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
count_cashback_gt_0 = df[df['cashback'] > 0].shape[0]
print(f"Total count where cashback > 0: {count_cashback_gt_0}")
count_cashback_eq_0 = df[df['cashback'] == 0].shape[0]
print(f"Total count where cashback = 0: {count_cashback_eq_0}")
count_coupon_yes = df[df['Coupon USed?'] == 'Yes'].shape[0]
print(f"Total count where Coupon Used? = 'Yes': {count_coupon_yes}")
count_coupon_no = df[df['Coupon USed?'] == 'No'].shape[0]
print(f"Total count where Coupon Used? = 'No': {count_coupon_no}")
labels = ['Cashback > 0', 'Cashback = 0', 'Coupon USed = Yes', 'Coupon USed = No']
values = [count_cashback_gt_0, count_cashback_eq_0, count_coupon_yes, count_coupon_no]
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['lightgreen', 'lightcoral', 'skyblue', 'salmon'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Counts of Different Conditions')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'booking_date' to datetime format
df['booking_date'] = pd.to_datetime(df['booking_date'])

# Sort the DataFrame by 'buyer_id' and 'booking_date'
df = df.sort_values(by=['buyer_id', 'booking_date'])

# Initialize a summary dictionary for retention
retention_summary = {'Same Supplier': 0, 'Different Supplier': 0}

# Group data by 'buyer_id' and calculate retention
for buyer_id, group in df.groupby('buyer_id'):
    suppliers = group['supplier_id'].unique()
    if len(suppliers) > 1:
        retention_summary['Different Supplier'] += 1
    else:
        retention_summary['Same Supplier'] += 1

# Prepare data for plotting
labels = list(retention_summary.keys())
values = list(retention_summary.values())

# Plotting the retention data using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['skyblue', 'salmon'])  # Changed to plt.bar()

# Add labels and title
plt.xlabel('Retention Type')
plt.ylabel('Number of Customers')
plt.title('Customer Retention by Supplier')

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
